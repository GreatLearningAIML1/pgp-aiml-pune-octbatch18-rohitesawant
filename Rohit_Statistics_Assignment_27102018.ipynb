```python
# Below code is to use Pandas to read the file
# The code then saves the file into a Dataframe called as laliga

# Initiate Pandas Library
import pandas as pd
import numpy as np

# Read laliga csv file in the dataframe and display
laliga = pd.read_csv('laliga.csv',header=1)
```

### Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (3 points)

```python
# Replace all dashes (-) with zeros (0)
laliga.replace('-',0,inplace=True)
# Read first 15 lines to check the results
laliga.head()
```

### Print all the teams which have started playing between 1930-1980. (5 points)

### Print the list of teams which came Top 5 in terms of points (3 points)

```python
# Points column is read as string because of the - in the file and hence converted to Numeric
laliga['Points']=laliga['Points'].apply(pd.to_numeric)
# Sorted by Points Column in Descending order and then printed the top 5 teams
laliga.sort_values(by=['Points'],ascending=False).head(5)[['Team','GoalsFor','GoalsAgainst']]
```

### Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (7 points)

```python
# Below Function should return the name of the team and goal differences
def Goal_diff_count():
    laliga[['GoalsFor','GoalsAgainst']]=laliga[['GoalsFor','GoalsAgainst']].apply(pd.to_numeric)
    laliga['GoalDiff']=laliga['GoalsFor']-laliga['GoalsAgainst']
    return laliga
# Assign Output of Function to a temp variable    
laliga_diff = Goal_diff_count()
```

```python
# Maximum difference of goals
max(laliga['GoalDiff'])
```

```python
# Minimum difference of goals
min(laliga['GoalDiff'])
```

### Create a new column with name “Winning Percent” and append it to the data set (6 points)
### Percentage of Winning = (GamesWon / GamesPlayed)*100
### If there are any numerical error, replace it with 0%
### Print the top 5 teams which has the highest Winning percentage

```python
# Conversion to Numeric to do Calculation
laliga[['GamesWon','GamesPlayed']]=laliga[['GamesWon','GamesPlayed']].apply(pd.to_numeric)
# Formula to calculate the Winning percentage with error handling for divide by 0
laliga['WinPercentage'] = np.where(laliga.GamesPlayed > 0,((laliga['GamesWon']/laliga['GamesPlayed'])*100),0)
# Print the top 5 teams by winning percentage
laliga.sort_values(by=['WinPercentage'],ascending=False).head(5)[['Team','WinPercentage']]
```
