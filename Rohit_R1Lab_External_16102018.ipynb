### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
iris.head(2)
```

```python
# Define variable for number of Columns
column_count=range(len(iris.columns))
# For each Column do the operation
for i in column_count:  
    # Do the Operations only if the column is a number
    if iris.iloc[:,i].dtype == np.float64:
        x=iris.iloc[:,i].max()
        y=iris.iloc[:,i].min()
        # Range Calculation 
        r = x - y
        # Column Name
        c = iris.columns[i]
        # Print Output in desired format
        print('Range of column ' + c,r)
        
```
